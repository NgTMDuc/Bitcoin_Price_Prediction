```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from datetime import datetime
```

```python
df1 = pd.read_csv('D:\\HUST\\20212\\Machine Learning\\Bitcoin_Price_Prediction\\data\\binance-BTCUSDT-1m.csv')
```

```python
df1
```

```python
feature = df1.columns
NA_ = np.array(list(x for x in feature if x.find('N/A') != -1))
NA_
```

```python
df1.shape
```

```python
def group_by(df, t):
    start_index = df.shape[0] - df.shape[0]//t * t
    newDf = pd.DataFrame(columns  = feature)
    for i in range(start_index,df.shape[0],t):
        tempdf = df[i: i+t]
        timestamp, opn = tempdf[['Timestamp', 'Open']].values[0] 
        high = tempdf['High'].max()
        low = tempdf['Low'].min()
        close, timeEnd = tempdf[['Close','Timestamp End']].values[-1]
        NA = tempdf[NA_].sum()
        row = [timestamp,opn,high,low,close,NA[0],timeEnd,NA[1], NA[2],NA[3], NA[4], NA[5], tempdf['Time_UTC_Start'].values[0]]
        new_df = pd.DataFrame([row], columns = feature)
        newDf = pd.concat([newDf, new_df], ignore_index = True)
    return newDf
```

```python
def label(df):
    length = df.shape[0]
    print(length)
    op = df['Open'].values.reshape(length,1)
    cl = df['Close'].values.reshape(length,1 )
    v = np.ones((length, 1)) * (cl < op)
    df['label'] = v
    return df
```

```python
df3 = group_by(df1, 15)
df4 = group_by(df1, 60)
df3 = label(df3)
df4 = label(df4)
df3.shape
```
