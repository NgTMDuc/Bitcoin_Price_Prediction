```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from datetime import datetime
```

```python
df1 = pd.read_csv('.\\data\\binance-BTCUSDT-1m.csv')
```

```python
df1
```

```python
def group(df,t):
    int_num = df.shape[0] //t
    start_index = df.shape[0] -int_num *t
    newDf = df[start_index:]
    index = np.ones(newDf.shape[0])
    for i in range(int_num):
        start = i * t
        end = (i + 1) * t
        index[start: end] *= i
    newDf['index'] = index
    return newDf
        
```

```python
df2 = group(df1,15)
```

```python
df2
```

```python
def label(df):
    grouped_single  = df.groupby('index').agg({'Timestamp': 'min', 'Open': 'first', 'High' : 'max', 'Low' : 'min', 'Close' : 'last',
                                              'N/A' : 'sum', 'Timestamp End' : 'max', 'N/A.1' : 'sum', 'N/A.2' : 'sum', 'N/A.3' : 'sum',
                                              'N/A.4' : 'sum', 'N/A.5' : 'sum', 'Time_UTC_Start': 'first'}).reset_index()
    length = grouped_single.shape[0]
    op = grouped_single['Open'].values.reshape(length,1)
    cl = grouped_single['Close'].values.reshape(length,1)
    l = np.ones((length, 1)) * (cl < op)
    grouped_single['label'] = l
    return grouped_single
```

```python
df3 = group(df1, 15)
df4 = group(df1, 60)
df3 = label(df3)
df4 = label(df4)
```

```python
df3
```
