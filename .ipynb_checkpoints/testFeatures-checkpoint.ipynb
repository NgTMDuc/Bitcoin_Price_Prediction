## Data preparation

```python
# Dataframe with 15 minutes between each row
def label(path, t=15):
    df = pd.read_csv(path)
    df = df[df.shape[0] % t:]
    df2 = df.groupby(np.arange(len(df))//t, axis=0).agg({'Timestamp': 'min', 'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last',
                                'N/A': 'sum', 'Timestamp End': 'max', 'N/A.1': 'sum', 'N/A.2': 'sum', 'N/A.3': 'sum',
                                'N/A.4': 'sum', 'N/A.5': 'sum', 'Time_UTC_Start': 'first'})
    df2.rename(columns = {'N/A': 'Volumn', 'N/A.1': 'Quote_asset_volume', 'N/A.2': 'Number_of_trades',
                         'N/A.3': 'Taker_buy_base_asset_volume', 'N/A.4': 'Taker_buy_quote_asset_volume'}, inplace = True)
    length = df2.shape[0]
    op = df2['Open'].values.reshape(length,1)
    cl = df2['Close'].values.reshape(length,1 )
    v = np.ones((length, 1)) * (op < cl)
    df2['label'] = v
    df2.index = df2["Timestamp"]
    df2.drop(columns=["Timestamp", "N/A.5"], inplace=True)
    return df2

df_15min = label("binance-BTCUSDT-1m.csv")
```

```python
# Create columns
df["Avg_price"] = df["Quote_asset_volume"]/df["Volumn"]

df.dropna(inplace=True)
c1 = pd.DataFrame({
    "Diff1": df["Close"] - df["Close"].shift(1),
    "Zero": [0 for i in np.arange(len(df))]
})
c2 = pd.DataFrame({
    "Diff2": df["Close"] - df["Close"].shift(1),
    "Zero": [0 for i in np.arange(len(df))]
})
df["Up"] = c1.max(axis=1)
df["Down"] = c2.min(axis=1)
df["Difference"] = df["Up"] + df["Down"]

df["Diff_Avg_price"] = df["Avg_price"] - df["Avg_price"].shift(1)
df["Diff_Quote_asset_volume"] = df["Quote_asset_volume"] - df["Quote_asset_volume"].shift(1)
df["Diff_Taker_buy_base_asset_volume"] = df["Taker_buy_base_asset_volume"] - df["Taker_buy_base_asset_volume"].shift(1)


for i in [5,10,20,50]:
    df["SMA%i"%(i)] = df["Close"].rolling(i, closed="left").mean() # closed="left" is used for avoiding leaked data
    df["SMAVolumn%i"%(i)] = df["Volumn"].rolling(i, closed="left").mean()
    df["SMAUp%i"%(i)] = df["Up"].rolling(i, closed="left").mean()
    df["SMADown%i"%(i)] = df["Down"].rolling(i, closed="left").mean()
    df.dropna(inplace=True)
    df["RS%i"%(i)] = df["SMAUp%i"%(i)] / (0.001 - df["SMADown%i"%(i)]) # 0.001 is used for avoiding zero division
    df["RSI%i"%(i)] = 100 - 100/(1 + df["RS%i"%(i)])
    
#     df["Diff_SMA%i"%(i)] = df["SMA%i"%(i)] - df["SMA%i"%(i)].shift(1)
#     df["Diff_SMAUp%i"%(i)] = df["SMAUp%i"%(i)] - df["SMAUp%i"%(i)].shift(1)
#     df["Diff_SMADown%i"%(i)] = df["SMADown%i"%(i)] - df["SMADown%i"%(i)].shift(1)
#     df["Diff_RS%i"%(i)] = df["RS%i"%(i)] - df["RS%i"%(i)].shift(1)
#     df["Diff_RSI%i"%(i)] = df["RSI%i"%(i)] - df["RSI%i"%(i)].shift(1)
    
    df.dropna(inplace=True)
    
#     df.drop(columns=["SMAUp%i"%(i),"SMADown%i"%(i), "SMA%i"%(i), "RS%i"%(i), "RSI%i"%(i)], inplace=True)

# # Drop columns
df.drop(columns=["Time_UTC_Start", "Timestamp End"], inplace=True)
# df.drop(columns=["Open", "High", "Low", "Avg_price", "Quote_asset_volume", "Taker_buy_base_asset_volume", "Avg_price"], inplace=True)

# # Save the index column of difference, we'll predict this feature
difference_idx = df.columns.get_loc("Difference")
```
